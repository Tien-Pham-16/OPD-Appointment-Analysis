```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from datetime import datetime as dt
```

# Data preparation and Cleaning

```python
#load data
df = pd.read_csv(r'C:\\Users\\Admin\\Downloads\\PROJECT AIH\\OPD appointment.csv')
df
```

```python
df.info()
```

```python
#remove unnessesary columns
df.drop(['Name', 'Registrator name', 'Doctor', 'Visit Code', 'Completed date', 'Pay status', 'OPD', 'ERO', 'ES', 'HC', 'Closed visit', 'Policy name']
,axis=1, inplace=True)
```

```python
df.info()
```

```python
#convert data type
df['PID'] = df['PID'].astype('str')
df['SMS'] = df['SMS'].astype('str')
df['Call'] = df['Call'].astype('str')
df['Appointment date'] = pd.to_datetime(df['Appointment date'])
df['Appointment time'] = pd.to_datetime(df['Appointment time'])
df['Arrival time'] = pd.to_datetime(df['Arrival time'])
```

```python
df.info()
df
```

```python
df.nunique()
```

There are 6195 customers come from 58 nations, 17 departments in the hospital

```python
#Create Delay time column by subtracting Appointment time column from the Arrival time column
df['Delay time'] = (df['Arrival time'] - df['Appointment time']).dt.total_seconds()/60
#extract only time
df['Appointment time'] = df['Appointment time'].dt.time
df['Arrival time'] = df['Arrival time'].dt.time
#rename column which is misspelled
df.rename(columns={'Deparment':'Department'},inplace = True)
df
```

# Handle null values

```python
#check null values
df.isnull().sum()
```

```python
#Replace null values at 'Arrival time' column and 'Delay time' column as '-100000'
df['Arrival time'].fillna('No show', inplace = True)
df['Delay time'].fillna('-100000', inplace = True)
df['Delay time'] = pd.to_numeric(df['Delay time'])
```

```python
df.isnull().any()
```

# Analysis and Visualization

```python
df
```

```python
#Visit distribution by Department
#Calculate base on PID
department_distribution = df.groupby('Department')['PID'].count().reset_index()
department_distribution.rename(columns = {'PID':'Number of visits'}, inplace = True)
department_distribution.sort_values('Number of visits', ascending=True, inplace = True)
department_distribution.reset_index(drop = True)
```

```python
plt.barh(department_distribution['Department'], department_distribution['Number of visits'])
plt.xlabel('Number of visits')
plt.ylabel('Department')
plt.title('Number of visits by Department')
plt.show()
```

There are 12,900 visits. The majority of customers come from the Pediatrics and the Obstetrics & Gynecology department, with a total of 8,298 visits, accounting for 64.32% of the total hospital visits. Next are the Endocrinology and ENT departments with 894 and 558 visits respectively

We will see Gender distribution to know how many percent Female patients and Male patients account for the total visits 

```python
#gender distribution 
gender_distribution = df.groupby('Gender')['PID'].nunique().reset_index()
gender_distribution.rename(columns = {'PID':'Amount'}, inplace = True)
gender_distribution
```

```python
#calculate percentage 
gender_distribution['Percentage'] = round(gender_distribution['Amount']/gender_distribution['Amount'].sum()*100,1)
gender_distribution
```

```python
#create pie chart
plt.pie(gender_distribution['Percentage'], autopct = '%1.1f%%', colors = ['pink', 'lightblue'], startangle= 45, explode = (0.025 , 0))
plt.title('Gender distribution')
plt.legend(['Female', 'Male'])
plt.show()
```

Based on the number of visits to each department, the Obstetrics & Gynecology department has the highest number of visits, followed by the Pediatrics department. Therefore, it is not surprising that Female customers account for 54.4% of the total visits, while males account for 45.6%

According to the hospital’s policy, any customer who arrives more than 15 minutes later than their scheduled appointment will be considered late and will be defaulted as not having an appointment. So we will calculate how many visit considered late based on the 'Delay time' column

```python
#filter visits with the delay time > 15 mins
late_visits = df[df['Delay time'] > 15]
late_visits
```

```python
#total late visits
late_visits['PID'].count()
```

```python
#total visits without no show
visits_without_noshow = df[df['Delay time'] != -100000]
visits_without_noshow['PID'].count()
```

We can see that there are 1393 over 10845 visits (without No show) that customers arrive later than their scheduled appointment. The rate of late visit is 12.8%

```python
department_late_visits = late_visits.groupby('Department')['Delay time'].count().reset_index()
department_late_visits.rename(columns={'Delay time':'Amount'}, inplace = True)
department_late_visits = department_late_visits.sort_values('Amount', ascending = False).reset_index(drop = True)
department_late_visits
```

We will see total visits (without No show) distribution by Department 

```python
department_distribution_without_noshow = visits_without_noshow.groupby('Department')['PID'].count().reset_index()
department_distribution_without_noshow.rename(columns = {'PID':'Number of visits'}, inplace = True)
department_distribution_without_noshow.sort_values('Number of visits', ascending=False, inplace = True)
department_distribution_without_noshow.reset_index(drop = True)
```

We'll calculate the rate of late visit by Department (without No show visits) 

```python
merged_departlate_departdistri = pd.merge(department_distribution_without_noshow,department_late_visits, on = 'Department')
merged_departlate_departdistri.rename(columns={'Amount':'Number of late visits'},inplace = True)
merged_departlate_departdistri['Rate (%)'] = round((merged_departlate_departdistri['Number of late visits']/merged_departlate_departdistri['Number of visits'])*100,1)
merged_departlate_departdistri.sort_values('Number of late visits',ascending = False,inplace = True)
merged_departlate_departdistri.reset_index(inplace = True,drop = True)
merged_departlate_departdistri
```

Arriving late will affect the hospital’s examination schedule and delay other patients who arrive on time, resulting in longer waiting times. This leads to a poor customer experience. 

        As we can see, Pediatrics has the highest number of late visits: 581/3755 visits, account for 15.5%. The Pediatrics department has the highest rate of late arrivals. The appointment and arrival times of patients in the Pediatrics department are influenced by many factors. Since the patients are children, but the accompanying individuals are their parents, the arrival time largely depends on the parents. In the Pediatrics department, arriving late for an appointment significantly affect the patient’s condition, as most of the children come for vaccinations and common childhood illnesses such as fever, cough, and nasal congestion.

        The department with the second highest number of late visits is Obstetrics and Gynecology (O&G), with 450 visits, accounting for 12.3% of the total visits to the O&G department. In the O&G department, arriving late for an appointment can significantly affect the entire process. Pregnancy is a long-term process, and expectant mothers will have to undergo various clinical tests such as Fasting Glucose tests, ultrasounds, fetal heart rate monitoring, etc. The O&G department has a very high number of visits, so not only does arriving late lead to delays and long waiting times for clinical tests, but it also affects the health of the expectant mother due to the requirement to be fasting for certain tests

        ENT has the third highest number of late visits with 97 out of 403 visits, accounting for 24.1%. This rate is relatively high and will greatly affect the customer experience because the waiting time for each ENT visit is quite long due to the ear, nose and throat issues that can range from simple to complex, involving many body systems. The complexity and diversity of these issues can sometimes lead to longer consultation times, in addition to the time required for procedures before and after endoscopy.

        The Endocrinology department has 63 late visits out of 658 visits, accounting for 9.6%. This is an acceptable rate for the Endocrinology department. Because the customers who visit the Endocrinology department are those with conditions related to hormones, thyroid glands, diabetes. These conditions often require long-term treatment because they are typically chronic and can significantly impact a person’s quality of life. Most patients will have to do blood tests, so arriving early or on time is necessary, as they need to be fasting for blood tests. Therefore, the rate of late visits in the Endocrinology department is not high.

See how the reminders work (SMS, Call)

```python
total_visits = 12900
sms_count = df.groupby('SMS')['PID'].count().reset_index()
sms_count.rename(columns={'PID':'Amount'},inplace=True)
call_count = df.groupby('Call')['PID'].count().reset_index()
call_count.rename(columns={'PID':'Amount'},inplace=True)
```

```python
sms_count['Percentage'] = sms_count['Amount']/total_visits*100
sms_count
```

```python
plt.pie(sms_count['Percentage'], autopct = '%1.0f%%', colors = ['LightGreen', 'Red'], startangle= 45)
plt.title('SMS Status')
plt.legend(['Done', 'Not yet'])
plt.show()
```

All visits were reminded by SMS

```python
call_count['Percentage'] = round(call_count['Amount']/total_visits*100,1)
call_count
```

```python
plt.pie(call_count['Percentage'], autopct = '%1.1f%%', colors = ['red', 'lightgreen'], startangle= 45, explode=(0.125,0))
plt.title('Call Status')
plt.legend(['Did not pick up', 'Done'])
plt.show()
```
